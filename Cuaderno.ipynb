```python
%matplotlib inline
from math import factorial
import matplotlib.pyplot as plt
from numpy import arange
ex = 0
aux = 0
exList = []
eaList = []
for x in range ( 0, 10 ) :
    aux = ex
    ex = ex + ( ( .5 ** x ) / factorial ( x ) )
    ea = ( ex - aux ) * 100 / ex
    exList.append ( ex )
    eaList.append ( ea )
    print ( "ex =", ex, "error porcentual =", ea, "%" )
plt.ylabel ( "e^x" )
plt.xlabel ( "iteraci√≥n" )
plt.title ( "e^x vs iteraci√≥n" )
plt.plot ( range ( 0, 10 ), exList )
fig = plt.figure()
plt.ylabel ( "error porcentual" )
plt.xlabel ( "iteraci√≥n" )
plt.title ( "error porcentual vs. iteraci√≥n" )
plt.plot ( range ( 0, 10 ), eaList )
fog = plt.figure()
```
