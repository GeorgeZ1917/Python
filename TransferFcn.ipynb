```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import control

sys = control.TransferFunction ( [ 10, 4 ], [ 1, 4, 4 ] )
response = control.step_response ( sys )
x = np.arange ( 0, len ( response [ 1 ] ) )
plt.xlabel ( "Time (seconds)" )
plt.ylabel ( "Amplitude" )
plt.plot ( x, response [ 1 ] )
```
